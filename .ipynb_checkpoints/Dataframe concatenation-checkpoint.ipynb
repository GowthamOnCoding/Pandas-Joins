# concatenation using the pd.concat() method

```python
import pandas as pd
week1=pd.read_csv("Restaurant - Week 1 Sales.csv")
week1.head()
```

```python
week2=pd.read_csv("Restaurant - Week 2 Sales.csv")
week2.head()
```

```python
len(week1)
```

```python
len(week2)
```

```python
#Concatenate week 1 and week 2 using concat method
pd.concat([week1,week2])
```

```python
pd.concat([week1,week2],ignore_index=True)
```

```python
#To knpw the source of data
df=pd.concat([week1,week2],keys=["week1","week2"])
df
```

```python
df.loc["week1"]
```

```python
df.loc["week2"]
```

```python
df.loc[("week1",200)]
```

```python
df.loc[("week2",99),"Food ID"]
```

```python
# The append() method
#Append is a method of dataframe and performs the same operation like concat
week2.append(week1)
```
